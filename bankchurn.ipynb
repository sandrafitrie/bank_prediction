# 1.Data perbsebaran male and female

```python
heh=data['Gender'].value_counts()
heh
```

```python
plt.style.use('fivethirtyeight')
```

```python
labels=['Perempuan','Laki-Laki']
```

```python
plt.pie(heh,labels=labels,wedgeprops={'edgecolor':'black'},startangle=90,autopct='%1.0f%%')
plt.title("Persebaran jumlah laki-laki dan perempuan")
plt.show()
```

# 2.Melihat Persebaran Status

```python
hm=data['Marital_Status'].value_counts()
hm
```

```python
labels=['Married','Single','Unknown','Divorced']
```

```python
plt.pie(hm,labels=labels,wedgeprops={'edgecolor':'black'},startangle=90,autopct='%1.0f%%')
plt.title('Status')
plt.show()
```

# 3.Mencari pendapatan yang paling dominan

```python
inc=data['Income_Category'].value_counts()
labels=['<$40K','$40K-$60K','$80K-$120K','$60K-$80K ','Unknown','$120K+']
```

```python
plt.pie(inc,labels=labels,wedgeprops={'edgecolor':'black'},startangle=90,autopct='%1.0f%%')
plt.show()
```

```python
plt.bar(labels,inc)
plt.show()
```
